```python
import keras
from keras import layers
import numpy as np

latent_dim = 100
height = 32
width = 32
channels = 3

#最適化アルゴリズム
#学習を安定させるために重みをclip＆weight decayを用いる
#識別者の学習ではRMSpropを用いたほうが良いとする事例もある
#またlrを非常に小さくする(0.00005以下)場合もある
#rmsp = keras.optimizers.RMSprop(lr=0.0008, clipvalue=1.0, decay=1e-8)
#識別者用＆敵対者用(DCGAN論文では両方Adam)
adam = keras.optimizers.Adam(lr=0.0002, beta_1=0.5, clipvalue=1.0, decay=1e-8)

#DC論文ではすべての重みを以下で初期化している
init = keras.initializers.RandomNormal(stddev=0.02)
```

```python
#CGAN生成ネットワーク
#潜在変数とラベルを掛け合わせたものをgeneratorの入力とする
generator_input = keras.Input(shape=(latent_dim,))
generator_label = keras.Input(shape=(1,), dtype='int32')

#Embeddingを用いてラベルを１００次元の密ベクトルに変換する
label_embedding = layers.Embedding(10, latent_dim, input_length=1)(generator_label)
label_embedding = layers.Flatten()(label_embedding)

#潜在変数とラベルを掛け合わせる
x = layers.Multiply()([label_embedding, generator_input])

#潜在変数を4x4 256-channels 特徴マップへ変換
x = layers.Dense(256 * 4 * 4, kernel_initializer=init, bias_initializer='zeros')(x)
x = layers.LeakyReLU(0.2)(x)
x = layers.Reshape((4, 4, 256))(x)

#4x4 256-channelsを8x8 128-channelsの特徴マップへDeconvolutionで変換
x = layers.Conv2DTranspose(128, 5, strides=2, padding='same', kernel_initializer=init, bias_initializer='zeros')(x)
x = layers.LeakyReLU(0.2)(x)

#8x8 128-channelsを16x16 64-channelsの特徴マップへDeconvolutionで変換
x = layers.Conv2DTranspose(64, 5, strides=2, padding='same', kernel_initializer=init, bias_initializer='zeros')(x)
x = layers.LeakyReLU(0.2)(x)

#16x16 64-channelsを32x32 32-channelsの特徴マップへDeconvolutionで変換
x = layers.Conv2DTranspose(32, 5, strides=2, padding='same', kernel_initializer=init, bias_initializer='zeros')(x)
x = layers.LeakyReLU(0.2)(x)

#32x32 32-channelsを３２x32 3-channelsの特徴マップへ変換
x = layers.Conv2D(channels, 5, activation='tanh', padding='same', kernel_initializer=init, bias_initializer='zeros')(x)
generator = keras.models.Model([generator_input, generator_label], x)
generator.summary()
```

```python
#CGAN識別ネットワーク
#ラベルと生成画像を結合したものを入力とする
discriminator_input = layers.Input(shape=(height, width, channels))
discriminator_label = layers.Input(shape=(1,), dtype='int32')

#Embeddingを用いてラベルを画像と同じサイズにReshape
label_embedding = layers.Embedding(10, height*width*channels, input_length=1)(discriminator_label)
label_embedding = layers.Flatten()(label_embedding)
label_embedding = layers.Reshape((height, width, channels))(label_embedding)

#ラベルと生成画像を結合
x = layers.Concatenate(axis=1)([discriminator_input, label_embedding])

x = layers.Conv2D(64, 5, strides=2,padding='same',kernel_initializer=init)(x)
x = layers.LeakyReLU(0.2)(x)
x = layers.Dropout(0.3)(x)
x = layers.Conv2D(128, 5, strides=2,padding='same')(x)
x = layers.LeakyReLU(0.2)(x)
x = layers.Dropout(0.3)(x)
x = layers.Conv2D(256, 5, strides=2,padding='same')(x)
x = layers.LeakyReLU(0.2)(x)
x = layers.Dropout(0.3)(x)
x = layers.GlobalAveragePooling2D()(x)
x = layers.Dense(1, activation='sigmoid')(x)

discriminator = keras.models.Model([discriminator_input, discriminator_label], x)
discriminator.summary()


discriminator_optimizer  = adam
discriminator.compile(optimizer=discriminator_optimizer, loss='binary_crossentropy')
```

```python
#CGAN敵対的ネットワーク
discriminator.trainable = False

gan_input = keras.Input(shape=(latent_dim,))
gan_label = keras.Input(shape=(1,), dtype='int32')
gan_output = discriminator([generator([gan_input, gan_label]), gan_label])
gan = keras.models.Model([gan_input, gan_label], gan_output)

gan_optimizer = adam
gan.compile(optimizer=gan_optimizer, loss='binary_crossentropy')
gan.summary()
```

```python
#画像表示用
def generateImage(epoch):
    print("epoch : %d" % (epoch+1))
    gan_labels = np.repeat(np.arange(10).reshape(-1, 1), 10, axis=0)
    #表示するサンプル数
    examples=100
    #潜在空間からサンプリング
    noise = np.random.normal(0, 1, size=[examples, latent_dim])
    #画像を生成
    generatedImages = generator.predict([noise, gan_labels])
    
    c = (generatedImages - np.min(generatedImages))/np.ptp(generatedImages)
    
    dim=(10, 10)
    plt.figure(figsize=(10, 10))
    for i in range(c.shape[0]):
        plt.subplot(dim[0], dim[1], i+1)
        plt.imshow(c[i].reshape(height, width, channels), interpolation='nearest', cmap='gray', vmax=1, vmin=0)
        plt.axis('off')
    plt.tight_layout()
```

```python
from keras.preprocessing import image
from keras.datasets import cifar10
from tqdm import tqdm
import matplotlib.pyplot as plt

#cifar10のダウンロード
(x_train, _), (x_test, y_test) = cifar10.load_data()

#各ピクセルが[0,1]の値を取るように正規化
#x_train = x_train.astype('float32') / 255.
#x_train = x_train.reshape(x_train.shape + (1,))
#x_test = x_test.astype('float32') / 255.
#x_test = x_test.reshape(x_test.shape + (1,))

#生成者の出力をtanhとしたので
#各ピクセルが[-1,1]の値を取るように正規化
x_train = (x_train.astype('float32')-127.5)/ 127.5
#x_train = x_train.reshape(x_train.shape + (1,))
x_test = (x_test.astype('float32')-127.5)/ 127.5
x_test = x_test.reshape(x_test.shape + (1,))

#DCGAN論文で使われているバッチ数を使う
batch_size = 128


epochs=1000
batchSize=batch_size
batchCount = x_train.shape[0] / batchSize
print ('Epochs:', epochs)
print ('Batch size:', batchSize)
print ('Batches per epoch:', batchCount)

# 学習のループ
for e in range(1, epochs+1):
    print('-'*15, 'Epoch %d' % e, '-'*15) 
    #for _ in tqdm(range(int(batchCount))): 
    for i in range(int(batchCount)):
        print("\r{0}% [{1}]".format(int(i/(batchCount-1)*100), "*" * int((i+1)/batchCount*50)), end="")
        # 潜在空間からランダムにサンプリング
        random_latent_vectors = np.random.normal(size=(batch_size, latent_dim))
        
        idx = np.random.randint(0, x_train.shape[0], size=batchSize)
        gan_labels = _[idx]
        discri_labels = np.concatenate([gan_labels, gan_labels])

        # データを生成する
        generated_images = generator.predict([random_latent_vectors, gan_labels])

        # 観測データと生成データを識別するための入力データをつくる
        real_images = x_train[idx]
        combined_images = np.concatenate([generated_images, real_images])

        # 観測データ(0)と生成データ(1)を識別するラベルをつくる
        labels = np.concatenate([np.ones((batch_size, 1)),
                             np.zeros((batch_size, 1))])
        
        # ランダムなノイズを加える
        labels *= 0.9

        #識別ネットワークの学習
        d_loss = discriminator.train_on_batch([combined_images, discri_labels], labels)

        #潜在空間からサンプリング
        random_latent_vectors = np.random.normal(size=(batch_size, latent_dim))

        # すべてリアルなデータだと騙すためのラベル(y=0)を生成
        misleading_targets = np.zeros((batch_size, 1))

        # 生成ネットワークの学習
        #100世代ごとに生成画像確認
        # min -y*log(D(G(z)))
        a_loss = gan.train_on_batch([random_latent_vectors, gan_labels], misleading_targets)
        if i==int(batchCount)-1 :
            print("\n")
            if e%100==0:
                print('discriminator loss: %s' % (d_loss))
                print('adversarial loss  : %s' % (a_loss))
                generateImage(e)
```
